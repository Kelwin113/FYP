#Data Preprocessing

```python
recipes_cleaned = recipes.copy()
reviews_cleaned = reviews.copy()
```

```python
# DATA MANIPULATION
# Columns dropping
columns_to_drop = ['DatePublished', 'RecipeIngredientQuantities', 'RecipeServings',
                   'RecipeYield','RecipeInstructions', 'AggregatedRating', 'ReviewCount']
recipes_cleaned = recipes_cleaned.drop(columns=columns_to_drop)

print('Before column dropping: ')
print(recipes.columns)
print('\nAfter column dropping: ')
print(recipes_cleaned.columns)
```

```python
# DATA CLEANING
# Removing missing values
recipes_cleaned = recipes_cleaned[~recipes_cleaned[['Images', 'Description', 'RecipeCategory']].isna().any(axis=1)]
reviews_cleaned = reviews_cleaned[~reviews_cleaned['Review'].isna()]
print(f"\nOriginal rows: {len(recipes)}, Rows after cleaning: {len(recipes_cleaned)}")
print(f"\nOriginal rows: {len(reviews)}, Rows after cleaning: {len(reviews_cleaned)}")
```

```python
# Remove Bath/Beauty recipe category rows
recipes_cleaned[recipes_cleaned['RecipeCategory'] == 'Bath/Beauty']["Name"]
```

```python
# Remove Bath/Beauty recipe category rows
recipes_cleaned = recipes_cleaned[recipes_cleaned['RecipeCategory'] != 'Bath/Beauty']

old = recipes[recipes['RecipeCategory'] == 'Bath/Beauty'].shape[0]
new = recipes_cleaned[recipes_cleaned['RecipeCategory'] == 'Bath/Beauty'].shape[0]
print(f"Count of Bath/Beauty food recipes in recipes: {old}")
print(f"\nCount of Bath/Beauty food recipes in recipes_cleaned: {new}")
```

```python
# Find recipeid in reviews not in recipes
missing_recipe_ids = set(reviews_cleaned['RecipeId'].unique()) - set(recipes_cleaned['RecipeId'].unique())

if len(missing_recipe_ids) > 0:
    # Filter reviews for rows where recipeid is not in recipes
    missing_recipe_reviews = reviews_cleaned[reviews_cleaned['RecipeId'].isin(missing_recipe_ids)].copy()

    if not missing_recipe_reviews.empty:
        print("RecipeId in reviews that is not in recipes:")
        print(missing_recipe_reviews.shape[0])

        # Remove rows from the original 'reviews' DataFrame
        reviews = reviews_cleaned[~reviews_cleaned['RecipeId'].isin(missing_recipe_ids)]
        print("\nRows with missing RecipeId removed from reviews.")
    else:
        print("No rows in reviews where recipeid is not in recipes.")
else:
    print("All recipe IDs in reviews are present in recipes.")
```

```python
# DATA TRANSFORMATION
# Convert DateSubmitted and DateModified columns to datetime format
reviews_cleaned['DateSubmitted'] = pd.to_datetime(reviews_cleaned['DateSubmitted'])
reviews_cleaned['DateModified'] = pd.to_datetime(reviews_cleaned['DateModified'])

print('Before conversion: ')
print(reviews[['DateSubmitted','DateModified']].head(4))
print('\nAfter conversion: ')
print(reviews_cleaned[['DateSubmitted','DateModified']].head(4))
```

```python
# Convert ingredients into a single string
recipes_cleaned["RecipeIngredientParts"] = recipes_cleaned["RecipeIngredientParts"].str.replace('c\(', '').str.replace('\"', '').str.replace('\)', '').str.split(', ')
recipes_cleaned["RecipeIngredientParts"] = recipes_cleaned["RecipeIngredientParts"].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)

print('Before conversion: ')
print(recipes["RecipeIngredientParts"].head(4))
print('\nAfter conversion: ')
print(recipes_cleaned["RecipeIngredientParts"].head(4))
```

```python
# Duration parsing and conversion
def parse_iso_duration(duration_str):
    if pd.isna(duration_str):
        return pd.NaT
    try:
        duration = pd.to_timedelta(duration_str)
        days = duration.days
        hours, remainder = divmod(duration.seconds, 3600)
        minutes, seconds = divmod(remainder, 60)
        result = f"{days} days {hours} hours {minutes} mins"
        return result
    except ValueError:
        return pd.NaT

# Apply the parsing function to 'CookTime', 'PrepTime', 'TotalTime'
for column in ['CookTime', 'PrepTime', 'TotalTime']:
    recipes_cleaned[column] = recipes_cleaned[column].apply(parse_iso_duration)

print('Before parsing: ')
print(recipes[['CookTime', 'PrepTime', 'TotalTime']].head(4))
print('\nAfter parsing: ')
print(recipes_cleaned[['CookTime', 'PrepTime', 'TotalTime']].head(4))
```

```python
# Extract the first image link containing ".jpg" or ".JPG" and remove surrounding quotes
recipes_cleaned["Images"] = recipes_cleaned["Images"].str.extract(r'\"(https[^\"]*\.(?:jpg|JPG)[^\"]*)\"', expand=False)

print('Before extraction: ')
print(recipes['Images'].head(4))
print('\nAfter extraction: ')
print(recipes_cleaned['Images'].head(4))
```

Add feature extraction

```python
# DATA NORMALIZATION
# Use z-score to normalize nutrtional values(+-3)
import numpy as np
from scipy.stats import zscore
columns_to_normalize = ['Calories', 'FatContent', 'SaturatedFatContent', 'CholesterolContent',
                         'SodiumContent', 'CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent']

z_scores = zscore(recipes_cleaned[columns_to_normalize])
recipes_cleaned[columns_to_normalize] = recipes_cleaned[columns_to_normalize][(np.abs(z_scores) < 3).all(axis=1)]

print('Before normalization: ')
for column in columns_to_normalize:
    print(f"{column} - {np.min(recipes[column]), np.max(recipes[column])}")

print('\nAfter normalization: ')
for column in columns_to_normalize:
    print(f"{column} - {np.min(recipes_cleaned[column]), np.max(recipes_cleaned[column])}")
```

```python
# Merge the two tables on the 'RecipeId' column and select specific columns
merged_reviews = pd.merge(reviews_cleaned, recipes_cleaned[['RecipeId', 'Name', 'Description']], on='RecipeId', how='left')

# Display column names
print("Initial reviews table columns:\n" + ', '.join(reviews.columns))
print("\nMerged reviews table columns:\n" + ', '.join(merged_reviews.columns))
reviews_cleaned = merged_reviews
reviews_cleaned = reviews_cleaned.dropna(subset=['Name'])
```
