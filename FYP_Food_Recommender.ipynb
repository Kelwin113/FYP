####3.1 Collaborative Filtering

####Model 1 - SVD User-Based Collaborate Filtering
- Purpose: Recommends items unrated by a user by identifying and suggesting products rated by other users with similar preferences and behavior.

- Algorithm: Singular Value Decomposition from Surprise


```python
from surprise import Dataset, Reader, SVD

rated_reviews = reviews_train[reviews_train['AuthorId'] == selected_author_id]
unrated_reviews = reviews_train[reviews_train['AuthorId'] != selected_author_id]
svd_df = reviews_train[reviews_train['Rating'] > 0]

# Load data into Surprise dataset
reader = Reader(rating_scale=(1, 5))
svd_train_data = Dataset.load_from_df(svd_df[['AuthorId', 'RecipeId', 'Rating']], reader)

# Build the full trainset
trainset = svd_train_data.build_full_trainset()

# Build and train the SVD model for collaborative filtering
svd_user_model = SVD(n_factors=50, lr_all=0.002, reg_all=0.02, n_epochs=30)
svd_user_model.fit(trainset)

# Get the list of all unique items (RecipeId) that the user has rated
rated_items = rated_reviews['RecipeId'].unique()

# Get the items that the user hasn't rated
unrated_items = unrated_reviews['RecipeId'].unique()

# Predict ratings for the unrated items
svd_user_ratings = [(selected_author_id, RecipeId,
                     svd_user_model.predict(selected_author_id, RecipeId).est) for RecipeId in unrated_items]

# Sort the predicted ratings in descending order
svd_user_ratings.sort(key=lambda x: x[2], reverse=True)

# Get the top N recommendations
svd_result = svd_user_ratings[:top_n]

# Print the top recommended items
print(f"Top {top_n} Recommended Food Items for User {selected_author_id} Using SVD Collaborative Filtering:")
for AuthorId, RecipeId, predicted_rating in svd_result:
    matched_recipe = recipes.loc[recipes['RecipeId'] == int(RecipeId)]
    recommended_recipe_name = matched_recipe.iloc[0]['Name']
    print(f"{RecipeId} - {recommended_recipe_name}")
```

####Model 2 - KNN User-Based Collaborate Filtering

- Algorithm: K-nearest Neighbors from Surprise


```python
from surprise import Dataset, Reader, KNNBasic

# Select all reviews for the given author
rated_reviews = reviews[reviews['AuthorId'] == selected_author_id]

# Sample the remaining reviews randomly
unrated_reviews = sampled_reviews[sampled_reviews['AuthorId'] != selected_author_id]

# Concatenate the two dataframes
knn_df = pd.concat([rated_reviews, unrated_reviews], ignore_index=True)
knn_df = knn_df[knn_df['Rating'] > 0]

# Load data into Surprise dataset
reader = Reader(rating_scale=(1, 5))
knn_train_data = Dataset.load_from_df(knn_df[['AuthorId', 'RecipeId', 'Rating']], reader)

# Build the full trainset
trainset = knn_train_data.build_full_trainset()

# Build and train the KNNBasic model for user-based collaborative filtering
sim_options = {
    'k': 30,
    'min_k': 1,
    'name': 'cosine',  # Use cosine similarity for user-based
    'user_based': True  # Set to True for user-based collaborative filtering
}

knn_user_model = KNNBasic(sim_options=sim_options)
knn_user_model.fit(trainset)

# Get the list of all unique items (RecipeId) that the user has rated
rated_items = rated_reviews['RecipeId'].unique()

# Get the items that the user hasn't rated
unrated_items = unrated_reviews['RecipeId'].unique()

# Predict ratings for the unrated items
knn_user_ratings = [
    (selected_author_id, RecipeId, knn_user_model.predict(selected_author_id, RecipeId).est)
    for RecipeId in unrated_items
]

# Sort the predicted ratings in descending order
knn_user_ratings.sort(key=lambda x: x[2], reverse=True)

# Get the top N recommendations
knn_result = knn_user_ratings[:top_n]

# Print the top recommended items
print(f"Top {top_n} Recommended Food Items for User {selected_author_id} Using KNN User-Based Collaborative Filtering:")
for AuthorId, RecipeId, predicted_rating in knn_result:
    matched_recipe = recipes.loc[recipes['RecipeId'] == int(RecipeId)]
    recommended_recipe_name = matched_recipe.iloc[0]['Name']
    print(f"{RecipeId} - {recommended_recipe_name}")
```

####Model 3 - TFIDF Vectorizer
- Purpose: Recommends items based on the characteristics or features of items. It considers the content or attributes of items and recommends items that are similar in content to what the user has liked or interacted with.

- Technique: TF-IDF (Term Frequency-Inverse Document Frequency) vectorization and cosine similarity from Scikit-learn

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
import numpy as np

# Reset index to avoid KeyError
sampled_reviews = sampled_reviews.reset_index(drop=True)

# Drop unnecessary columns
content_data = sampled_reviews[['AuthorId', 'RecipeId', 'Description', 'Review', 'Name', 'Rating']]

# Drop rows with missing values in the 'Review' and 'Name' columns
content_data = content_data.dropna(subset=['Name'])

# Concatenate 'Review' and 'Rating' columns
content_data['combined_text'] = content_data['Review'].fillna('') + ' ' + content_data['Description'].fillna('') + ' ' + content_data['Rating'].astype(str)

# Create a TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(content_data['combined_text'])

# Calculate cosine similarity between reviews
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

def get_top_similar_recipes(author_id, user_profile=user_profile_table, cosine_sim=cosine_sim, content_data=content_data, N=10):
    # Get the index of the selected author in the content_data
    idx = content_data[content_data['AuthorId'] == author_id].index

    if not idx.empty:
        idx = idx[0]

        # Calculate cosine similarity between reviews
        sim_scores = list(enumerate(cosine_sim[idx]))

        # Personalize results based on user ratings
        sim_scores = [
            (i, score * (1 + content_data.iloc[i]['Rating'] - user_profile[user_profile['RecipeId'] ==
            content_data.iloc[i]['RecipeId']]['Rating'].values[0]))
            for i, score in sim_scores
        ]
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

        # Filter out recipes already reviewed by the selected author
        author_reviews = content_data[content_data['AuthorId'] == author_id]['RecipeId'].tolist()
        sim_scores = [(i, score) for i, score in sim_scores if content_data.iloc[i]['RecipeId'] not in author_reviews]

        # Take top N similar recipes
        sim_scores = sim_scores[:N]
        recipe_indices = [i for i, _ in sim_scores]
        return content_data.iloc[recipe_indices]
    else:
        print(f"Author with ID {author_id} not found.")
        return pd.DataFrame()

# Use the updated function to get top similar recipes
tfidf_result = get_top_similar_recipes(selected_author_id)

# Print the top 10 personalized recommended recipes
print(f"Top 10 Personalized Recommended Food for Author ID {selected_author_id} using TFIDF Vectorizer:")
for index, row in tfidf_result.iterrows():
    print(f"{row['RecipeId']} - {row['Name']}")
```

```python
# Create a table to store user past interaction for evaluation
# Filter rows for selected_author_id
user_interaction = content_data[content_data['AuthorId'] == selected_author_id].copy()

# Drop rows with missing values in the 'Review' and 'Name' columns
user_interaction = user_interaction.dropna(subset=['Name'])

# Concatenate 'Review' and 'Rating' columns
user_interaction['combined_text'] = user_interaction['Review'].fillna('') + ' ' + user_interaction['Description'].fillna('') + ' ' + user_interaction['Rating'].astype(str)
```

```python
# Function to calculate cosine similarity between user interaction and recommended results
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def calculate_cosine_similarity(user_interaction_text, recommended_text):
    combined_text_all = pd.concat([user_interaction_text, recommended_text], axis=0)

    # Create a TF-IDF Vectorizer
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix_all = tfidf_vectorizer.fit_transform(combined_text_all)

    # Calculate cosine similarity between user interaction and recommended results
    cosine_sim_user_recommendations = cosine_similarity(tfidf_matrix_all[:1], tfidf_matrix_all[1:]).flatten()

    return cosine_sim_user_recommendations

# Get the top 10 personalized recommended recipes
recommended_recipes = get_top_similar_recipes(selected_author_id, N=10)

# Print the top 10 personalized recommended recipes with cosine similarity
print(f"Cosine Similarity between User Interaction and Recommended Results for {selected_author_id} Using TFIDF Vectorizer:")
average_cosine_similarity = 0.0  # Initialize for calculating the average

for index, row in recommended_recipes.iterrows():
    recipe_text = row['combined_text']
    cosine_similarities = calculate_cosine_similarity(user_interaction['combined_text'], pd.Series([recipe_text]))

    # Display the recommendation with its cosine similarity score
    print(f"{row['RecipeId']} - {row['Name']} ({round(cosine_similarities[0], 4)})")

    # Accumulate the cosine similarity score for calculating the average
    average_cosine_similarity += cosine_similarities[0]

# Calculate and display the average cosine similarity
average_cosine_similarity /= len(recommended_recipes)
print(f"\nAverage Cosine Similarity for TFIDF Vectorizer Recommended Items: {round(average_cosine_similarity, 4)}")
```

####Model 4 - Word2Vec
- Purpose: Recommends items based on the characteristics or features of items. It considers the content or attributes of items and recommends items that are similar in content to what the user has liked or interacted with.

- Technique: Word2Vec from gensim library

```python
import nltk
from sklearn.metrics.pairwise import linear_kernel
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import numpy as np

# Reset index to avoid KeyError
sampled_reviews = sampled_reviews.reset_index(drop=True)

# Drop unnecessary columns
content_data = sampled_reviews[['AuthorId', 'RecipeId', 'Description', 'Review', 'Name', 'Rating']]

# Drop rows with missing values in the 'Name' columns
content_data = content_data.dropna(subset=['Name'])

# Tokenize the combined text into words
content_data['combined_text'] = content_data['Review'].fillna('') + ' ' + content_data['Description'].fillna('') + ' ' + content_data['Rating'].astype(str)
content_data['tokenized_text'] = content_data['combined_text'].apply(word_tokenize)

# Function to average word vectors for a document
def average_word_vectors(words, model, vocabulary, num_features):
    feature_vector = np.zeros((num_features,), dtype="float32")
    nwords = 0.
    for word in words:
        if word in vocabulary:
            nwords = nwords + 1.
            feature_vector = np.add(feature_vector, model.wv[word])
    if nwords:
        feature_vector = np.divide(feature_vector, nwords)

    return feature_vector

# Train Word2Vec model
word2vec_model = Word2Vec(sentences=content_data['tokenized_text'], vector_size=100, window=5, min_count=1, workers=4)

# Create a feature matrix by averaging word vectors for each document
vocabulary = set(word2vec_model.wv.index_to_key)
word2vec_num_features = 100  # Change this to match the vector_size parameter in Word2Vec

content_data['word2vec_features'] = content_data['tokenized_text'].apply(
    lambda x: average_word_vectors(x, word2vec_model, vocabulary, word2vec_num_features)
)

# Create a DataFrame with the feature vectors
word2vec_features_df = pd.DataFrame(content_data['word2vec_features'].tolist(), index=content_data.index)

# Concatenate the word2vec_features_df with other relevant columns
content_data = pd.concat([content_data, word2vec_features_df], axis=1)

# Calculate cosine similarity between word2vec features
cosine_sim_word2vec = linear_kernel(word2vec_features_df, word2vec_features_df)

def get_top_similar_recipes_word2vec(author_id, user_profile=user_profile_table, cosine_sim=cosine_sim_word2vec,
                                     content_data=content_data, N=10):
    # Get the index of the selected author in the content_data
    idx = content_data[content_data['AuthorId'] == author_id].index

    if not idx.empty:
        idx = idx[0]

        # Calculate cosine similarity between Word2Vec features
        sim_scores = list(enumerate(cosine_sim[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

        # Filter out recipes already reviewed by the selected author
        author_reviews = content_data[content_data['AuthorId'] == author_id]['RecipeId'].tolist()
        sim_scores = [(i, score) for i, score in sim_scores if content_data.iloc[i]['RecipeId'] not in author_reviews]

        # Weigh similarity scores based on user's ratings
        sim_scores = [
            (i, score * (1 + content_data.iloc[i]['Rating'] - user_profile[user_profile['RecipeId'] ==
                                                                      content_data.iloc[i]['RecipeId']]['Rating'].values[0]))
            for i, score in sim_scores
        ]
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

        # Create a DataFrame with the recommendations
        recommended_recipes = pd.DataFrame(columns=content_data.columns)
        seen_recipe_ids = set()

        for i, _ in sim_scores:
            recipe_id = content_data.iloc[i]['RecipeId']

            # Check if the recipe is not already in the recommended_recipes list
            if recipe_id not in seen_recipe_ids:
                recommended_recipes = pd.concat([recommended_recipes, content_data.iloc[[i]]])
                seen_recipe_ids.add(recipe_id)

            # Break if N recommendations are reached
            if len(recommended_recipes) >= N:
                break

        return recommended_recipes.head(N)
    else:
        print(f"Author with ID {author_id} not found.")
        return pd.DataFrame()

# Use the updated function to get top similar recipes with Word2Vec
word2vec_result = get_top_similar_recipes_word2vec(selected_author_id)

# Print the top 10 personalized recommended recipes with Word2Vec
print(f"Top 10 Personalized Recommended Food for Author ID {selected_author_id} using Word2Vec:")
for index, row in word2vec_result.iterrows():
    print(f"{row['RecipeId']} - {row['Name']}")
```

```python
import pandas as pd
from gensim.models import FastText
from nltk.tokenize import word_tokenize
import numpy as np
from sklearn.metrics.pairwise import linear_kernel

# Reset index to avoid KeyError
sampled_reviews = sampled_reviews.reset_index(drop=True)

# Drop unnecessary columns
content_data = sampled_reviews[['AuthorId', 'RecipeId', 'Description', 'Review', 'Name', 'Rating']]

# Drop rows with missing values in the 'Name' columns
content_data = content_data.dropna(subset=['Name'])

# Tokenize the combined text into words
content_data['combined_text'] = content_data['Review'].fillna('') + ' ' + content_data['Description'].fillna('') + ' ' + content_data['Rating'].astype(str)
content_data['tokenized_text'] = content_data['combined_text'].apply(word_tokenize)

# Function to average word vectors for a document
def average_word_vectors(words, model, num_features):
    feature_vector = np.zeros((num_features,), dtype="float32")
    nwords = 0.

    for word in words:
        nwords = nwords + 1.
        feature_vector = np.add(feature_vector, model.wv[word])

    if nwords:
        feature_vector = np.divide(feature_vector, nwords)

    return feature_vector

# Create a feature matrix by averaging word vectors for each document
fasttext_num_features = 100

# Train FastText model
fasttext_model = FastText(sentences=content_data['tokenized_text'], vector_size=100, window=5, min_count=1, workers=4)

content_data['fasttext_features'] = content_data['tokenized_text'].apply(
    lambda x: average_word_vectors(x, fasttext_model, fasttext_num_features)
)

# Create a DataFrame with the feature vectors
fasttext_features_df = pd.DataFrame(content_data['fasttext_features'].tolist(), index=content_data.index)

# Concatenate the fasttext_features_df with other relevant columns
content_data = pd.concat([content_data, fasttext_features_df], axis=1)

# Calculate cosine similarity between FastText features
cosine_sim_fasttext = linear_kernel(fasttext_features_df, fasttext_features_df)

def get_top_similar_recipes_fasttext(author_id, user_profile=user_profile_table, cosine_sim=cosine_sim_fasttext, content_data=content_data, N=10):
    idx = content_data[content_data['AuthorId'] == author_id].index
    if not idx.empty:
        idx = idx[0]
        sim_scores = list(enumerate(cosine_sim[idx]))

        # Weigh similarity scores based on user's ratings
        sim_scores = [
            (i, score * (1 + content_data.iloc[i]['Rating'] - user_profile[user_profile['RecipeId'] ==
                                                                      content_data.iloc[i]['RecipeId']]['Rating'].values[0]))
            for i, score in sim_scores
        ]

        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

        author_reviews = content_data[content_data['AuthorId'] == author_id]['RecipeId'].tolist()
        sim_scores = [(i, score) for i, score in sim_scores if content_data.iloc[i]['RecipeId'] not in author_reviews]
        sim_scores = sim_scores[:N]
        recipe_indices = [i for i, _ in sim_scores]

        # Create a DataFrame with the recommendations using pd.concat
        recommended_recipes = pd.DataFrame(columns=content_data.columns)
        seen_recipe_ids = set()

        for i in recipe_indices:
            recipe_id = content_data.iloc[i]['RecipeId']

            # Check if the recipe is not already in the recommended_recipes list
            if recipe_id not in seen_recipe_ids:
                recommended_recipes = pd.concat([recommended_recipes, content_data.iloc[[i]]])
                seen_recipe_ids.add(recipe_id)

        return recommended_recipes.head(N)
    else:
        print(f"Author with ID {author_id} not found.")
        return pd.DataFrame()

# Use the updated function to get top similar recipes with FastText
fasttext_result = get_top_similar_recipes_fasttext(selected_author_id)

# Print the top 10 personalized recommended recipes with FastText
print(f"Top 10 Personalized Recommended Food for Author ID {selected_author_id} using FastText:")
for index, row in fasttext_result.iterrows():
    print(f"{row['RecipeId']} - {row['Name']}")
```
